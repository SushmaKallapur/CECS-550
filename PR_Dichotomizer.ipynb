```python
# Load the Pandas libraries with alias 'pd'
import statistics
import math
import pandas as pd

train_data = pd.read_csv('./HW2-TrainData.csv', usecols=[0, 1, 2, 3], header=None, skiprows=[0])
test_data = pd.read_csv('./HW2-TestData.csv', usecols=[0, 1, 2, 3], header=None, skiprows=[0])

def Average(lst):
    return sum(lst) / len(lst)
def probability(x,u,s):
    number = (x-u)/s
    # print("number", number)
    ex_data = -0.5 * math.pow(number,2)
    # print("ex_data", ex_data)
    uni_variate = (0.4 * math.exp(ex_data))/s
    # print("uni_variate", uni_variate)
    return uni_variate
def getStatistics(trainDataSet, testDataSet, w, x):
    # rslt_train_data = trainDataSet.loc[trainDataSet['Load'] == w]
    # next(trainDataSet, None)
    train_data_list_x = trainDataSet[x]
    # test data
    rslt_test_data = testDataSet.loc[testDataSet[3] == w]
    test_data_list_x = rslt_test_data[x]

    mean = Average(train_data_list_x)
    # print("mean", mean, x, w)
    standard_deviation = statistics.pstdev(train_data_list_x)
    # print("SD", standard_deviation, x, w)
    list = 'probability_list_' + str(x) + '_' + w
    # print(list)
    list = []

    for i in test_data_list_x:
        answer = probability(i, mean, standard_deviation)
        answer = answer * 0.5

        list.append(answer)
    print("probability", list)
    return list
def compareTwoLists(list1,list2):
    # compare elements in two lists
    index = 0
    compare_list = []
    for i in list1:
        if i > list2[index]:
            compare_list.append('w1')
        else:
            compare_list.append('w2')
        index = index + 1
    return compare_list

def combineTwoLists(list1,list2):
    # compare elements in two lists
    index = 0
    combine_list = []
    for i in list1:
        product_x1_x2 = i * list2[index]
        combine_list.append(product_x1_x2)
        index = index + 1
    return combine_list

def getErrorRate(list1,list2,s,e):
    # compare elements in two lists
    index = 0
    compare_list = []
    for i in list1:
        if i == list2[index]:
            s = s + 1
        else:
            e = e + 1
        index = index + 1
        if index >= len(list2):
            break
    sum_of_success_and_error = s+e
    error_percentage = (e/sum_of_success_and_error) * 100
    return error_percentage




def calculateErrorRateOfX1():
    print("****************X1******************")
    list_w1_x1 = getStatistics(train_data, test_data,  'w1', 0)
    list_w2_x1 = getStatistics(train_data, test_data, 'w2', 0)

    compared_list = compareTwoLists(list_w1_x1, list_w2_x1)
    print(compared_list)

    test_data_list = test_data[3]

    print(test_data_list)

    success_rate = 0
    error_rate = 0

    errors = getErrorRate(compared_list, test_data_list, success_rate, error_rate)
    print(errors)
def calculateErrorOfX1AndX2():
    print("****************X1***X2******************")
    list_w1_x1 = getStatistics(train_data, test_data, 'w1', 0)
    list_w2_x1 = getStatistics(train_data, test_data, 'w2', 0)
    list_w1_x2 = getStatistics(train_data, test_data, 'w1', 1)
    list_w2_x2 = getStatistics(train_data, test_data, 'w2', 1)


    combine_list_w1_x1_x2 = combineTwoLists(list_w1_x1, list_w1_x2)
    combine_list_w2_x1_x2 = combineTwoLists(list_w2_x1, list_w2_x2)
    compare_list = compareTwoLists(combine_list_w1_x1_x2, combine_list_w2_x1_x2)
    print(compare_list)

    test_data_list = test_data[3]

    print(test_data_list)

    success_rate = 0
    error_rate = 0

    errors = getErrorRate(compare_list, test_data_list, success_rate, error_rate)
    print(errors)



def calculateErrorOfX1AndX2AndX3():
    print("************X1****X2*******X3***********")
    list_w1_x1 = getStatistics(train_data, test_data, 'w1', 0)
    list_w2_x1 = getStatistics(train_data, test_data, 'w2', 0)
    list_w1_x2 = getStatistics(train_data, test_data, 'w1', 1)
    list_w2_x2 = getStatistics(train_data, test_data, 'w2', 1)
    list_w1_x3 = getStatistics(train_data, test_data, 'w1', 2)
    list_w2_x3 = getStatistics(train_data, test_data, 'w2', 2)

    combine_list_w1_x1_x2 = combineTwoLists(list_w1_x1, list_w1_x2)
    combine_list_w1_x1_x2_x3 = combineTwoLists(combine_list_w1_x1_x2, list_w1_x3)
    combine_list_w2_x1_x2 = combineTwoLists(list_w2_x1, list_w2_x2)
    combine_list_w2_x1_x2_x3 = combineTwoLists(combine_list_w2_x1_x2, list_w2_x3)

    compare_list = compareTwoLists(combine_list_w1_x1_x2_x3, combine_list_w2_x1_x2_x3)
    print(compare_list)

    test_data_list = test_data[3]

    print(test_data_list)

    success_rate = 0
    error_rate = 0

    errors = getErrorRate(compare_list, test_data_list, success_rate, error_rate)
    print(errors)


# var = rslt_df.loc[rslt_df['x1']]
# data = data.iloc[:data.loc[data.Load.str.contains('W2', na=False)].index[0]]
# print(mean)
# print(standard_deviation)
if __name__ == '__main__':
    calculateErrorRateOfX1()
    calculateErrorOfX1AndX2()
    calculateErrorOfX1AndX2AndX3()
    # train_data.head()
    # test_data.head()
```
